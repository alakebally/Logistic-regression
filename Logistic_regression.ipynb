**Identify categorical predictors**

```python
df.select_dtypes(include='object').columns
```

**Separate the target variable**

```python
X = df.drop('satisfaction', axis=1)
y = df['satisfaction']
```

**Encode the categorical predictors**

```python
X_encoded = pd.get_dummies(X, drop_first=True)
```

**Encode the target variable(satisfaction)**

```python
y = y.map({
    'dissatisfied': 0,
    'satisfied': 1
})
```

**Vertfy the result**

```python
X_encoded.head()
```

```python
X_encoded.info()
```

**Split data into training and testing sets**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

**Scale the predictors using StandardScalar**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Linear REgression Assumption**


 **Binary Outcome Variable**
The target variable (satisfaction) contains two categories: Satisfied and Dissatisfied. This satisfies the binary outcome requirement for Logistic Regression.

 **Missing Data**

The dataset was inspected for missing values. The only variable with missing observations was Arrival Delay in Minutes, and these missing values were imputed using the median due to the highly skewed distribution of the variable.

**Independent Observations**

Each row in the dataset represents an individual passenger survey response and is assumed to be independent of other observations.

**Feature Encoding**

Categorical variables were converted into numerical format using one-hot encoding, making them suitable for Logistic Regression.

**Feature Scaling**

Continuous predictors were standardized using StandardScaler to improve model convergence and optimization performance.

 **Linearity of the Log-Odds**

Logistic Regression assumes a linear relationship between continuous predictors and the log-odds of the outcome variable. A formal Box-Tidwell test was not conducted as part of this project; therefore, this assumption was not explicitly verified and should be investigated in future work.

**Fit the model**

```python

from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(
    max_iter=1000,
    random_state=42
)

log_reg.fit(X_train_scaled, y_train)
```

**Generate the predicted probabilities for the "satisfied" class**

```python
y_pred = log_reg.predict(X_test_scaled)

y_prob = log_reg.predict_proba(X_test_scaled)[:, 1]
#Display the first 10 predicted probabilites
pd.Series(y_prob).head(10)
```

**Generate the confusion matrix**

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

**Confusion Matrix Plot**

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()

plt.title("Confusion Matrix")
plt.show()
```

**Visualizing the Confusion Matrix**
The confusion matrix provides a detailed breakdown of correct and incorrect classifications, allowing evaluation of true positives, true negatives, false positives, and false negatives.

**Calculate Precision and Recall**

```python
from sklearn.metrics import precision_score, recall_score

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Precision:", precision)
print("Recall:", recall)
```

**Generate a classification Report**

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

**Model Evaluation Summary**

| Metric    | Value |
| --------- | ----: |
| Accuracy  |   83% |
| Precision | 84.4% |
| Recall    | 84.3% |
| F1-Score  |   84% |


**Interpretation**

The model achieved balanced performance across both classes. Precision and recall were both approximately 84%, indicating that the model effectively identifies satisfied passengers while maintaining a relatively low false positive rate.

**Evaluate the F1-Score**

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)

print("F1-Score:", f1)
```

**Interpretation**


The model achieved an F1-score of approximately 0.84, indicating a strong balance between precision and recall.

**Limitations**


1. **Model assumptions**

Logistic Regression assumes a linear relationship between predictors and the log-odds of satisfaction, which may not fully capture complex customer behaviors.

2. **Limited feature set**

The model only uses variables available in the dataset. Other factors influencing satisfaction, such as ticket price, weather disruptions, or airport conditions, were not included.

3. **Historical data**

The model was trained on historical observations and may not fully reflect future changes in customer preferences.

4. **Correlation vs causation**

The model identifies associations with satisfaction but does not prove that a variable directly causes satisfaction.
